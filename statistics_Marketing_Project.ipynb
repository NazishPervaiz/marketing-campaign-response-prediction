```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, normaltest, probplot

# Features to check
features_to_check = ['Income', 'Age', 'NumWebVisitsMonth']

# Create plots and perform tests
for feature in features_to_check:
    data = df[feature].dropna()
    
    # Visual: Histogram + Q-Q Plot
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    sns.histplot(data, kde=True, bins=30)
    plt.title(f'Histogram: {feature}')
    
    plt.subplot(1, 2, 2)
    probplot(data, dist="norm", plot=plt)
    plt.title(f'Q-Q Plot: {feature}')
    
    plt.tight_layout()
    plt.show()

    # Statistical Tests
    shapiro_stat, shapiro_p = shapiro(data)
    normaltest_stat, normaltest_p = normaltest(data)

    print(f"\n Normality Test Results for: {feature}")
    print(f"Shapiro-Wilk Test:   stat={shapiro_stat:.4f}, p={shapiro_p:.4g}")
    print(f"D’Agostino-Pearson:  stat={normaltest_stat:.4f}, p={normaltest_p:.4g}")
    if shapiro_p < 0.05 or normaltest_p < 0.05:
        print("Likely NOT normally distributed")
    else:
        print("Likely normally distributed")
```

```python
import pandas as pd

# Columns to check for outliers
columns_to_check = [
    'Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
    'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
    'NumStorePurchases', 'NumWebVisitsMonth', 'Age', 'MntTotal',
    'AcceptedCmpOverall', 'Customer_Days', 'MntRegularProds'
]

# Store results here
outlier_summary = {}

# Loop through each column to detect outliers using IQR
for col in columns_to_check:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    
    # Count unique outlier values
    outlier_counts = outliers.value_counts().sort_index()
    
    # Save in dictionary
    outlier_summary[col] = outlier_counts

# Display outlier summary as tables
for col, counts in outlier_summary.items():
    if not counts.empty:
        print(f"\n🔎 Outliers in column: {col}")
        display(pd.DataFrame({col: counts.index, 'Count': counts.values}))
    else:
        print(f"\n✅ No outliers detected in column: {col}")
```

```python
import pandas as pd
from scipy.stats import mannwhitneyu, chi2_contingency

# Define target
target = 'Response'

# All numeric features (non-normal)
numeric_non_normal = [
    'Income', 'Recency', 'Age', 'Customer_Days', 'NumWebVisitsMonth',
    'NumStorePurchases','NumDealsPurchases','NumWebPurchases','NumCatalogPurchases',
    'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds','MntTotal','MntRegularProds'
]

# Categorical features
categorical_features = [
    'education_2n Cycle', 'education_Basic', 'education_Graduation', 'education_Master', 'education_PhD',
    'marital_Single', 'marital_Divorced', 'marital_Widow', 'marital_Married', 'marital_Together', 
    'accepted_campaign', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'Kidhome', 'Teenhome',
    'AcceptedCmp4', 'AcceptedCmp5'
]

# Store results
mannwhitney_results = []
chi2_results = []

# Mann-Whitney U Test for all non-normal numeric features
for feature in numeric_non_normal:
    group0 = df[df[target] == 0][feature]
    group1 = df[df[target] == 1][feature]
    stat, p = mannwhitneyu(group0, group1, alternative='two-sided')
    mannwhitney_results.append({'Feature': feature, 'U Statistic': stat, 'p-value': p})

# Chi-Square Test for categorical features
for feature in categorical_features:
    contingency = pd.crosstab(df[feature], df[target])
    stat, p, dof, expected = chi2_contingency(contingency)
    chi2_results.append({
        'Feature': feature,
        'Chi2 Statistic': stat,
        'p-value': p,
        'Degrees of Freedom': dof
    })

# Convert to DataFrames and round
mannwhitney_df = pd.DataFrame(mannwhitney_results).round(5)
chi2_df = pd.DataFrame(chi2_results).round(5)

# Display results
print("\n Mann-Whitney U Test Results")
display(mannwhitney_df)

print("\n Chi-Square Test Results")
display(chi2_df)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Assume your dataset is stored in a DataFrame called df
df_transformed = df.copy()

# 1. Log-transform all skewed features
log_features = [
    'Income', 'Age', 'Customer_Days', 'NumDealsPurchases',
    'NumWebPurchases', 'NumCatalogPurchases',
    'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds'
]

for col in log_features:
    df_transformed[col] = np.log1p(df_transformed[col])

# 2. Scale 'Recency' using MinMaxScaler
minmax = MinMaxScaler()
df_transformed['Recency'] = minmax.fit_transform(df_transformed[['Recency']])

# 3. Apply StandardScaler to all remaining features (including log-transformed ones)
standard_features = log_features + [
    'NumWebVisitsMonth', 'NumStorePurchases'
]
scaler = StandardScaler()
df_transformed[standard_features] = scaler.fit_transform(df_transformed[standard_features])
```

```python
leaky_features = [
    'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5',
    'AcceptedCmpOverall'
]

X = df_transformed.drop(columns=['Response'] + leaky_features)
y = df_transformed['Response']

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,         # 20% test split
    random_state=42,       # For reproducibility
    stratify=y             # Preserve class distribution
)
```

```python
print(X_train.select_dtypes(include=['object', 'category']).columns)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Define numeric features (that need scaling)
#numeric_features = [f for f in features_to_check if f in X_train_bal.columns]
numeric_features = log_features + ['Recency', 'NumWebVisitsMonth', 'NumStorePurchases']

# Fit scaler on balanced training data numeric features and transform
X_train_bal_scaled = X_train_bal.copy()
X_train_bal_scaled[numeric_features] = scaler.fit_transform(X_train_bal[numeric_features])

# Transform test data numeric features using the same scaler (do NOT fit again)
X_test_scaled = X_test.copy()
X_test_scaled[numeric_features] = scaler.transform(X_test[numeric_features])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced', random_state=42, max_iter=2000)
model.fit(X_train_bal_scaled, y_train_bal)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

y_pred = model.predict(X_test_scaled)
y_proba = model.predict_proba(X_test_scaled)[:, 1]

print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear', 'lbfgs'],
    'penalty': ['l2']
}

grid_search = GridSearchCV(
    estimator=LogisticRegression(class_weight='balanced', random_state=42, max_iter=1000),
    param_grid=param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)

# Fit on balanced, scaled training data
grid_search.fit(X_train_bal_scaled, y_train_bal)

print("Best hyperparameters:", grid_search.best_params_)

# Use best model
best_model = grid_search.best_estimator_

# Evaluate on test set
y_pred = best_model.predict(X_test_scaled)
y_proba = best_model.predict_proba(X_test_scaled)[:, 1]

print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

coefs = pd.Series(best_model.coef_[0], index=X.columns)
coefs.sort_values().plot(kind='barh', figsize=(10, 6))
plt.title("Logistic Regression Coefficients")
plt.xlabel("Coefficient Value")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

rf = RandomForestClassifier(class_weight='balanced', random_state=42)

rf_grid = GridSearchCV(
    estimator=rf,
    param_grid=rf_params,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)

rf_grid.fit(X_train_bal_scaled, y_train_bal)
print("RF Best Params:", rf_grid.best_params_)
```
